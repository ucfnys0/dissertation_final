```python
import pandas as pd
import os
def combine(file1,file2):
    data1=pd.read_excel(file1)
    data2 = pd.read_excel(file2)
    data=pd.merge(data1,data2,on="LSOA code")
    return data
path=r"..\..\result"
file1=os.path.join(path,"kmeans_result_predict.xlsx")
file2=os.path.join(path,"predict.xlsx")
data=combine(file1,file2)
data
```

```python
data.to_excel(os.path.join(path,"kmeans_result_prediction.xlsx"))
```
