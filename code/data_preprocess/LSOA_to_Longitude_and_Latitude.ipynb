```python
import pandas as pd
import numpy as np
import os
path="..\..\dataset\screened_crime"
def get_location(path):
    location=pd.DataFrame(columns=["LSOA code","Longitude","Latitude"])
    location.set_index("LSOA code",inplace=True)

    for dirpath, dirnames, filenames in os.walk(path):
        for filename in filenames:
            crime_data=pd.read_excel(os.path.join(dirpath, filename))
            crime_data.set_index("LSOA code",inplace=True)
            for i, row in crime_data.iterrows():
                if i not in location.index:
                    location.loc[i]=[row["Longitude"],row["Latitude"]]


    return location
location=get_location(path)

location
```

```python
location.to_excel(r"..\..\dataset\LSOA_code.xlsx")
```
