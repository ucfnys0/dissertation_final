```python
import folium
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel(r"..\..\dataset\kmeans_predict_for_draw_correct.xlsx")
data
```

```python
limit =200
data = data.sample(limit)
```

```python
m = folium.Map(location=[51.5, 0], zoom_start=12)
for index, row in data.iterrows():
    word=" <b>Urban safety factor</b>: "+str(np.round(row["final score"],2))+"<br>"+"<b>Safety degree</b>: "+str(row["final_label"])+"<br>"+"<b>Crime rate</b>:"+str(row["crime_score"])+"<br>"+"<b>Retail convenience</b>:"+str(row["retail_score"])+"<br>"+"<b>Surroundings</b>:"+str(row["environment_score"])+"<br>"+"<b>Medical degree</b>:"+str(row["health_score"])+"<br>"+"<b>AQI</b>:"+str(row["air_label"])+"<br>"+"<b>IMD</b>:"+str(row["poor_score"])+".\n"
    folium.Marker(
        location=[row["Latitude"],row["Longitude"]],
        popup=folium.Popup(word,max_width=450),
        icon=None
    ).add_to(m)
m
```

```python
m.save('final.html')     
```
